````python
import joblib

import nltk
from nltk.tokenize import word_tokenize
from nltk.corpus import stopwords
from nltk.stem import WordNetLemmatizer

nltk.download('punkt', quiet=True)
nltk.download('stopwords', quiet=True)
nltk.download('wordnet', quiet=True)

# Initialize once
stop_words = set(stopwords.words('english'))
lemmatizer = WordNetLemmatizer()
````

````python
from score import score
````

````python
# label mapper
label_map = {0: "ham", 1: "spam"}

# Preprocessing Function
def preprocess_text(text: str) -> List[str]:
    """
    Preprocess raw SMS text into cleaned tokens, using following steps:
    - Tokenization
    - Lowercasing
    - Stopword removal
    - Lemmatization
    - Remove non-alphabetic tokens
    """

    if not isinstance(text, str):
        raise ValueError("Text must be a string")
    
    tokens = word_tokenize(text) # Tokenization
    tokens = [token.lower() for token in tokens]  # Convert to lowercase first
    tokens = [token for token in tokens if token.isalpha()] # Keep only alphabetic tokens (removes punctuation & numbers)
    tokens = [token for token in tokens if token not in stop_words] # Remove stopwords
    tokens = [lemmatizer.lemmatize(token) for token in tokens] # Lemmatization

    return tokens
````

````python
model = joblib.load("BestModel_NaiveBayes.pkl")
print("Model loaded")
````

````python
text = "Congratulations! You won a free lottery ticket!!!"
tokens = preprocess_text(text)
print(tokens)
````

````python
prediction, propensity = score(
    "Congratulations! You won a free lottery ticket!!!",
    model,
    threshold=0.5
)

print("Message:", text)
print("prediction:", prediction)
print("Prediction:", label_map[prediction])
print("Propensity:", round(propensity, 4))
````
