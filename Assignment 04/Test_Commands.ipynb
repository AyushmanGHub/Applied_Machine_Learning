### Pre-Commit Hook

A **Git hook** is a script that runs automatically when certain Git events occur. Git provides several hooks such as `pre-commit`, `post-commit`, and `pre-push` that allow developers to automate tasks during the development workflow.

A **pre-commit hook** runs **before a commit is finalized**. It allows you to check or validate the code before it is added to the repository.

### Purpose of the Pre-Commit Hook
**Pre-commit hook is used to run automated tests before allowing a commit**. This ensures that only code that `passes all tests` is committed to the repository.

When user runs:

```bash
git commit -m "message"
```

Git first executes the `pre-commit` script. The script runs `pytest` to execute the tests.

* If the tests **pass**, the commit proceeds.
* If the tests **fail**, the commit is **aborted**.

It prevents committing broken or failing code and ensures code quality and reliability. It also automates testing before code is added to the repository.
> Acts as a basic form of **continuous integration (CI)** on the local machine.
